### Correlation
Comparision between two variables as to if they are related or not, is measured using the strength of correlation, also known as correlation co-efficient
```
 + 1 - Directly proportional, if one variable increases then the other also increases
 - 1 - Indirectly proportional, if one variable increases then the other decreases
   0 - No relation
```

````python
# Heat Map

plt.figure(figsize=(12, 8))
sns.heatmap(boston_df.corr(),
            cmap="Blues",
            xticklabels=boston_df.corr().columns.values,
            yticklabels=boston_df.corr().columns.values,
            annot=True,
            linecolor='white')
plt.xticks(rotation=50)
plt.show()
````
